```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/ds.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["SupplierID" ,"SupplierAddress" ,"SupplierName"],axis=1,inplace=True)
```

```python
df.drop(columns=["CreditCard","CarModel","CarMaker","SupplierContactDetails","ProductID", "CustomerID","CustomerName","Gender","JobTitle","PhoneNumber","EmailAddress","City","CountryCode","CustomerAddress","OrderDate", "OrderID","ShipDate","PostalCode"],axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.CarColor=le.fit_transform(df.CarColor)
df.Country=le.fit_transform(df.Country)
df.State=le.fit_transform(df.State)
df.ShipMode=le.fit_transform(df.ShipMode)
df.Shipping=le.fit_transform(df.Shipping)
df.CreditCardType=le.fit_transform(df.CreditCardType)
df.CustomerFeedback=le.fit_transform(df.CustomerFeedback)
```

```python
df.head()
```

```python
x=df.drop(columns="CarPrice",axis=1)
x.head()
```

```python
y=df[["CarPrice"]]
y[0:5]
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
x_scaled=pd.DataFrame(scale.fit_transform(x),columns = x.columns)
y_scaled= pd.DataFrame(scale.fit_transform(y),columns = y.columns)
```

```python
x_scaled.head()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
score=r2_score(y_pred,y_test)
score
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

reg_model=Sequential()
reg_model.add(Dense(9,activation="relu"))
reg_model.add(Dense(128,activation="relu"))
reg_model.add(Dense(1,activation="linear"))
```

```python

reg_model.compile(optimizer="adam",loss="mse",metrics=["accuracy"])
```

```python

reg_model.fit(x_train,y_train,epochs=10,batch_size=5,validation_data=(x_test,y_test))
```
