```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("startups.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.State=le.fit_transform(df.State)
```

```python
name=["R&D Spend","Administration","Marketing Spend","State","Profit"]
```

```python
import matplotlib.pyplot as plt
for i,j in enumerate(name):
    plt.figure(figsize=(12,15))
    plt.subplot(4,5,i+1)
    sns.boxplot(df[j])
```

```python
x=df.drop(columns='Profit',axis=1)
x.head()
```

```python
y=df.Profit
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Build an ANN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
reg_model=Sequential()
```

```python
#ADDING THE INPUT LAYER
reg_model.add(Dense(4,activation="relu"))
#ADDING ONE HIDDEN LAYER
reg_model.add(Dense(64,activation="relu"))
#ADDING 2ND HIDDEN LAYER
reg_model.add(Dense(32,activation="relu"))
#ADDING 3RD HIDDEN LAYER
reg_model.add(Dense(16,activation="relu"))
#ADDING OUTPUT LAYER
reg_model.add(Dense(1,activation="linear"))
```

```python
#COMPLING THE OUTPUT
reg_model.compile(optimizer="adam",loss="mse")
```

```python
#FITTING THE MODELANN
reg_model.fit(x_train,y_train,batch_size=2,epochs=10)
```

```python
y_pred=reg_model.predict(x_test)
```

```python
y_pred
```

```python
#PREDICTIONG THE MODEL FOR A NEW ITEM
reg_model.predict([[324234,4545645,234234,2]])
```

```python
# Build an ANN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
classification_model=Sequential()
```

```python
#ADDING THE INPUT LAYER
classification_model.add(Dense(4,activation="relu"))
#ADDING ONE HIDDEN LAYER
classification_model.add(Dense(64,activation="relu"))
#ADDING 2ND HIDDEN LAYER
classification_model.add(Dense(32,activation="relu"))
#ADDING 3RD HIDDEN LAYER
classification_model.add(Dense(16,activation="relu"))
#ADDING OUTPUT LAYER
classification_model.add(Dense(1,activation="sigmoid"))
```

```python
#COMPLING THE OUTPUT
classification_model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
classification_model.fit(x_train,y_train,epochs=10,batch_size=3,validation_data=(x_test,y_test))
```

```python
y_pred=classification_model.predict(x_test)
```

```python
classification_model.predict([[324234,4545645,234234,2]])
```
