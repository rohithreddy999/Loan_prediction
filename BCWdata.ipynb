```python
import pandas as pd
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("BCWdata.csv")  
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop(columns=["Unnamed: 32","id"],axis=1,inplace=True)
```

```python
df
```

```python
df.diagnosis.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.diagnosis=le.fit_transform(df.diagnosis)
df.head()
```

```python
x=df.drop(columns="diagnosis",axis=1)
y=df.diagnosis
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#initializing the object
classification_model=Sequential()
```

```python
#building an ann model for classification
#input layer  
classification_model.add(Dense(30,activation="relu"))
#hidden layers------->
classification_model.add(Dense(64,activation="relu"))
classification_model.add(Dense(36,activation="relu"))
classification_model.add(Dense(16,activation="relu"))
classification_model.add(Dense(10,activation="relu"))
#output layer
classification_model.add(Dense(1,activation="sigmoid"))
```

```python
classification_model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
classification_model.fit(x_train,y_train,batch_size=8,epochs=400,validation_data=(x_test,y_test))
```

```python
y_pred=classification_model.predict(x_test)
```
