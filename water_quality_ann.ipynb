```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("/content/water_potability.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
 df.loc[(df['ph'].isna() == True), 'ph'] = df['ph'].mean()
 df.loc[(df['Sulfate'].isna() == True), 'Sulfate'] = df['Sulfate'].mean()
 df.loc[(df['Trihalomethanes'].isna() == True), 'Trihalomethanes'] = df['Trihalomethanes'].mean()
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=['Potability'],axis = 1)

     
```

```python
y=df.Potability
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
x_scaled= pd.DataFrame(scale.fit_transform(x),columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
reg_model=Sequential()
reg_model.add(Dense(9,activation="relu"))
reg_model.add(Dense(128,activation="relu"))
reg_model.add(Dense(1,activation="linear"))
```

```python
reg_model.compile(optimizer="adam",loss="mse",metrics=["accuracy"])
```

```python
reg_model.fit(x_train,y_train,epochs=400,batch_size=75,validation_data=(x_test,y_test))
```
